```python
import pandas as pd
```

```python
jan_2020 = pd.read_csv("JC-202001-citibike-tripdata.csv")
feb_2020 = pd.read_csv("JC-202002-citibike-tripdata.csv")
march_2020 = pd.read_csv("JC-202003-citibike-tripdata.csv")
april_2020 = pd.read_csv("JC-202004-citibike-tripdata.csv")
may_2020 = pd.read_csv("JC-202005-citibike-tripdata.csv")
june_2020 = pd.read_csv("JC-202006-citibike-tripdata.csv")
```

```python
all_city = jan_2020.append(feb_2020)
all_city = all_city.append(march_2020)
all_city = all_city.append(april_2020)
all_city = all_city.append(may_2020)
all_city = all_city.append(june_2020)
```

```python
all_city
```

```python
all_city['gender']=all_city["gender"].replace([0],"unknown")
```

```python
all_city['gender']=all_city["gender"].replace([1],"male")
```

```python
all_city['gender']=all_city["gender"].replace([2],"female")
```

```python
all_city
```

```python
all_city.to_csv("city_bike_data.csv", index=False, header=True)
```
